```python
class Student:
    School = " NED University"
    Department = " BCIT "
    Batch = " 2nd YEAR "
    
    def __init__(self,std_id,subjects):
        self.id = std_id
        self.sub_names = subjects
        
    def inp(self):
        self.id = int(input("\n\n Enter your Student ID : "))
        self.sub_names=[ ]
        list_num = int(input(" How Many subjects you want to be registered in : "))
        print(" Enter Subject Names: ")
        
        for i in range(list_num):
            subj=input(" ")
            self.sub_names.append( subj )
        
    def calculate(self):
        if len(self.sub_names) == 1:
            print("Total Fees : ", 5000)
        elif len(self.sub_names) == 2:
            print("Total Fees : ", 10000)
        else:
            print("Total Fees : ", 15000)
        
    def display(self):
        print( "\n\n Student ID = ", self.id )
        print( " Subjects Selected = ",self.sub_names)

print(" \t Subjects Offered In Spring Semester 2020 : \n 1. OOP \n 2.DAA \n 3.CG ")


s1= Student(0,0)
s2= Student(0,0)
s3= Student(0,0)


s1.inp()
s2.inp()
s3.inp()

print(" \n\n\t\t SCHOOL NAME = ",s1.School )
print(" \t\t DEPARTMENT NAME = ", s1.Department )
print(" \t\t BATCH NO. = " ,s1.Batch )
s1.display()
s1.calculate()

print(" \n\n\t\t SCHOOL NAME = ",s2.School )
print(" \t\t DEPARTMENT NAME = ",s2.Department )
print(" \t\t BATCH NO. = " ,s2.Batch )
s2.display()
s2.calculate()

print(" \n\n\t\t SCHOOL NAME = ",s3.School )
print(" \t\t DEPARTMENT NAME = ", s3.Department )
print(" \t\t BATCH NO. = " ,s3.Batch )
s3.display()
s3.calculate()

```
